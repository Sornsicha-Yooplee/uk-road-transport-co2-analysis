```python
import os
import pandas as pd
import numpy as np

DATA_DIR = r"D:\UWE CLASS\CSCT master\Master_project\Data"

BREAKDOWN_FILE = os.path.join(DATA_DIR, "region_year_vehicle_co2e_breakdown.csv")

df_bd = pd.read_csv(BREAKDOWN_FILE)
print("shape:", df_bd.shape)
print("columns:", list(df_bd.columns))
display(df_bd.head(10))
```

```python
year_col = next((c for c in df_bd.columns if c.lower() == "year" or "year" in c.lower()), None)
region_col = next((c for c in df_bd.columns if "region" in c.lower() and "code" not in c.lower() and "id" not in c.lower()), None)
veh_col = next((c for c in df_bd.columns if "vehicle" in c.lower() or "type" in c.lower()), None)
em_col = next((c for c in df_bd.columns if "co2" in c.lower() and ("ton" in c.lower() or "emiss" in c.lower())), None)

print("Detected -> year:", year_col, "| region:", region_col, "| vehicle:", veh_col, "| emissions:", em_col)

if year_col:
    print("Year range:", df_bd[year_col].min(), "to", df_bd[year_col].max())
if veh_col:
    print("Vehicle types (sample):", df_bd[veh_col].dropna().unique()[:15])
```
