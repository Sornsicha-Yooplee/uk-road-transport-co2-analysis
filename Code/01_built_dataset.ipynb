```python
import pandas as pd

PATH = r"D:\UWE CLASS\CSCT master\Master_project\Data\traffic_2015_2024.csv"
df = pd.read_csv(PATH)

df.head()
```

```python
vehicle_cols = [
    "two_wheeled_motor_vehicles",
    "cars_and_taxis",
    "buses_and_coaches",
    "LGVs",
    "all_HGVs"
]

# เช็คว่าคอลัมน์มีจริงครบ
missing = [c for c in vehicle_cols if c not in df.columns]
print("Missing vehicle cols:", missing)
```

```python
panel = df.melt(
    id_vars=["year", "region_name", "region_ons_code"],
    value_vars=vehicle_cols,
    var_name="vehicle_type",
    value_name="activity_miles"
)

panel.head()


```

```python
MILES_TO_KM = 1.60934
panel["activity_km"] = panel["activity_miles"] * MILES_TO_KM

panel[["year","region_name","vehicle_type","activity_miles","activity_km"]].head()
```

```python
total_km = panel.groupby(["region_name", "year"])["activity_km"].sum().rename("total_km")

panel = panel.join(total_km, on=["region_name", "year"])
panel["share_activity"] = panel["activity_km"] / panel["total_km"]

panel.head()




```

```python
check = panel.groupby(["region_name","year"])["share_activity"].sum().reset_index()
check["diff_from_1"] = (check["share_activity"] - 1).abs()
check.sort_values("diff_from_1", ascending=False).head()
```

```python
OUT = r"D:\UWE CLASS\CSCT master\Master_project\Data\panel_activity_long.csv"
panel.to_csv(OUT, index=False)
print("Saved:", OUT)
```

```python
import pandas as pd

CF_PATH = r"D:\UWE CLASS\CSCT master\Master_project\Data\ghg-conversion-factors-2024-FlatFormat_v1_1.xlsx"

tmp = pd.read_excel(CF_PATH, sheet_name="Factors by Category", header=None)
header_row = tmp.index[tmp.iloc[:,0].astype(str).str.strip().eq("ID")][0]

cf = pd.read_excel(CF_PATH, sheet_name="Factors by Category", header=header_row)

print("Header row:", header_row)
print("Shape:", cf.shape)
cf.head()
```

```python
UOM_COL = "UOM"
GHG_COL = "GHG/Unit"
FACTOR_COL = "GHG Conversion Factor 2024"

L1 = "Level 1"
L2 = "Level 2"
L3 = "Level 3"
L4 = "Level 4"
CT = "Column Text"


kmf = cf[
    (cf[UOM_COL].astype(str).str.strip() == "km") &
    (cf[GHG_COL].astype(str).str.contains("kg CO2e", case=False, na=False))
].copy()

print("km factors shape:", kmf.shape)
kmf[[L1, L2, L3, CT, UOM_COL, GHG_COL, FACTOR_COL]].head(10)
```

```python
kmf_total = kmf[kmf[GHG_COL].astype(str).str.strip().eq("kg CO2e")].copy()
print("km factors (total kg CO2e) shape:", kmf_total.shape)
kmf_total[[L1, L2, L3, CT, UOM_COL, GHG_COL, FACTOR_COL]].head(15)
```

```python
def show_candidates(df, label, cond):
    sub = df.loc[cond, [L1, L2, L3, L4, CT, UOM_COL, GHG_COL, FACTOR_COL]].copy()
    print(f"\n--- {label}: candidates = {len(sub)} ---")
    display(sub.head(20))
    return sub
```

```python
car_cand = show_candidates(
    kmf_total,
    "CAR",
    kmf_total[L2].astype(str).str.contains("Cars", case=False, na=False)
)

lgv_cand = show_candidates(
    kmf_total,
    "VAN/LGV",
    kmf_total[L2].astype(str).str.contains("Vans", case=False, na=False)
)

hgv_cand = show_candidates(
    kmf_total,
    "HGV",
    kmf_total[L2].astype(str).str.contains("HGV", case=False, na=False)
)

moto_cand = show_candidates(
    kmf_total,
    "MOTORBIKE",
    kmf_total[L2].astype(str).str.contains("Motor", case=False, na=False)
)
```

```python
vehicle_map = {
    "cars_and_taxis": "car",
    "LGVs": "lgv",
    "all_HGVs": "hgv",
    "buses_and_coaches": "bus",
    "two_wheeled_motor_vehicles": "motorcycle",
}

panel["vehicle_std"] = panel["vehicle_type"].map(vehicle_map)
panel["vehicle_std"].isna().sum(), panel["vehicle_std"].unique()
```

```python
car_rows = kmf_total[
    (kmf_total[L1] == "Passenger vehicles") &
    (kmf_total[L2].astype(str).str.contains("Cars", case=False, na=False)) &
    (kmf_total[CT].astype(str).str.strip().eq("Unknown"))
].copy()

car_rows = car_rows[car_rows[FACTOR_COL].notna()]

print("car rows:", car_rows.shape)
display(car_rows[[L2, L3, CT, FACTOR_COL]].head(20))

ef_car = car_rows[FACTOR_COL].mean()
print("ef_car (kg CO2e/km):", ef_car)
```

```python
lgv_rows = kmf_total[
    (kmf_total[L1].astype(str).str.contains("Delivery", case=False, na=False)) &
    (kmf_total[L2].astype(str).str.contains("Vans", case=False, na=False)) &
    (kmf_total[CT].astype(str).str.strip().eq("Unknown"))
].copy()

lgv_rows = lgv_rows[lgv_rows[FACTOR_COL].notna()]

print("lgv rows:", lgv_rows.shape)
display(lgv_rows[[L2, L3, CT, FACTOR_COL]].head(20))

ef_lgv = lgv_rows[FACTOR_COL].mean()
print("ef_lgv (kg CO2e/km):", ef_lgv)
```

```python
moto_rows = kmf_total[
    kmf_total[L2].astype(str).str.contains("Motorbike|Motorcycle", case=False, na=False)
].copy()

moto_rows = moto_rows[moto_rows[FACTOR_COL].notna()]

print("motorcycle rows:", moto_rows.shape)
display(moto_rows[[L1, L2, L3, CT, FACTOR_COL]].head(30))

ef_moto = moto_rows[FACTOR_COL].mean()
print("ef_motorcycle (kg CO2e/km):", ef_moto)
```

```python
pkm = cf[
    (cf["UOM"].astype(str).str.strip() == "passenger.km") &
    (cf["GHG/Unit"].astype(str).str.strip() == "kg CO2e")
].copy()

bus_pkm = pkm[pkm["Level 2"].astype(str).str.contains("Bus|Coach", case=False, na=False)].copy()
bus_pkm = bus_pkm[bus_pkm["GHG Conversion Factor 2024"].notna()]

display(bus_pkm[["Level 1","Level 2","Level 3","Column Text","GHG Conversion Factor 2024"]].head(20))

# ===== สมมติฐาน (ปรับได้) =====
AVG_PASSENGERS_BUS = 10
# ==============================

ef_bus = float(bus_pkm["GHG Conversion Factor 2024"].mean() * AVG_PASSENGERS_BUS)
ef_bus
```

```python
ef = pd.DataFrame([
    ["car", ef_car],
    ["lgv", ef_lgv],
    ["hgv", ef_hgv],
    ["motorcycle", ef_moto],
    ["bus", ef_bus],
], columns=["vehicle_std", "ef_kgco2e_per_km"])

display(ef)

panel2 = panel.merge(ef, on="vehicle_std", how="left")

missing = panel2.loc[panel2["ef_kgco2e_per_km"].isna(), "vehicle_std"].unique()
print("Missing EF for:", missing)

panel2["co2e_kg"] = panel2["activity_km"] * panel2["ef_kgco2e_per_km"]
panel2["co2e_tonnes"] = panel2["co2e_kg"] / 1000

region_year = (
    panel2.groupby(["region_name","year"], as_index=False)["co2e_tonnes"]
          .sum()
          .rename(columns={"co2e_tonnes":"co2e_tonnes_total"})
)

OUT1 = r"D:\UWE CLASS\CSCT master\Master_project\Data\region_year_co2e.csv"
region_year.to_csv(OUT1, index=False)

OUT2 = r"D:\UWE CLASS\CSCT master\Master_project\Data\region_year_vehicle_co2e_breakdown.csv"
panel2[["region_name","year","vehicle_std","activity_km","share_activity","co2e_tonnes"]].to_csv(OUT2, index=False)

print("Saved:", OUT1)
print("Saved:", OUT2)
region_year.head()
```
